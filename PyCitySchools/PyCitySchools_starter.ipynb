# PyCity Schools Analysis

Charter schools outperformed district schools in terms of average math and reading, as well as student who  passed math and reading. Factors determining performance will require more exploration. 

Small and Medium sized schools (2000 or fewer students) outperformed larger schools in all areas. 

Students at schools with higher budgets ($645 per student or more) were outperformed by students at schools with smaller budgets.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on="school_name")
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools - use len() on the list of the unique school
school_count = len(school_data_complete["school_name"].unique())
school_count
```

```python
# Calculate the total number of students - count the number of student Ids
student_count = school_data_complete["Student ID"].count()
student_count
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score - take the mean of the math_score column from school data complete dataframe
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score - take the mean of the reading_score column from school data complete dataframe
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# use conditional to find all students who has a score of 70 or higher in math
passMathStudents = school_data_complete[school_data_complete["math_score"] >= 70]
#passMathStudents.head()
numPassMath = len(passMathStudents)
numPassMath
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
# (number of students who passed math / total number of students) * 100
percentagePassedMath = (numPassMath / student_count) * 100
percentagePassedMath
```

```python
# use conditional to find all students who has a score of 70 or higher in math
passReadingStudents = school_data_complete[school_data_complete["reading_score"] >= 70]
#passReadingStudents.head()
numPassReading = len(passReadingStudents)
numPassReading
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
# (number of students who passed reading / total number of students) * 100
passing_reading_percentage = (numPassReading / student_count) * 100
passing_reading_percentage
```

```python
# use conditional to find all students who has a score of 70 or higher in math and reading
passingStudents = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
#passingStudents.head()
numPassingStudents = len(passingStudents)
numPassingStudents
```

```python
# Use the following to calculate the percentage of students that passed math and reading
# (number of students who passed math and reading / total number of students) * 100
overall_passing_rate = (numPassingStudents / student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame - make a dictionary of lists
district_summary = pd.DataFrame(
    {
        "Total Schools": [school_count],
        "Total Students": [student_count],
        "Total Budget": [total_budget],
        "Average Math Score": [average_math_score],
        "Average Reading Score": [average_reading_score],
        "% Passing Math": [percentagePassedMath],
        "% Passing Reading": [passing_reading_percentage],
        "% Passing Reading": [passing_reading_percentage],
        "% Overall Passing": [overall_passing_rate],
    }
)

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("{:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.2f}%".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:.2f}%".format)

# Display the DataFrame
district_summary
```

```python
# calculate the average test scores
groupedschoolDF = school_data_complete.groupby("school_name")
# get the average reading scores
avgReadingScoreDF = groupedschoolDF.mean("reading_score")
#avgReadingScoreDF
# convert to a series by grabbing just the reading_score column
perSchoolReadingScores = avgReadingScoreDF["reading_score"]
perSchoolReadingScores
```

```python
# Calculate the number of students per school with math scores of 70 or higher from school_data_complete
schoolPassingMathDF = school_data_complete[school_data_complete["math_score"] >= 70]
schoolPassingMathDF
```

```python
# Calculate the number of students per school with reading scores of 70 or higher from school_data_complete
schoolPassingReadingDF = school_data_complete[school_data_complete["reading_score"] >= 70]
schoolPassingReadingDF
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
schoolPassingDF = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
schoolPassingDF
```

```python
# use a groupby to count the number of students to each school who passed math
perSchoolPassingMathDF = schoolPassingMathDF.groupby("school_name")
perSchoolPassingMath = perSchoolPassingMathDF["student_name"].count()
perSchoolPassingMath
```

```python
# use a groupby to count the number of students to each school who passed reading
perSchoolPassingReadingDF = schoolPassingReadingDF.groupby("school_name")
perSchoolPassingReading = perSchoolPassingReadingDF["student_name"].count()
perSchoolPassingReading
```

```python
perSchoolPassingDF = schoolPassingDF.groupby("school_name")
perSchoolPassing = perSchoolPassingDF["student_name"].count()
perSchoolPassing
```

```python
# Use the provided code to calculate the passing rates
perSchoolPassingMathPCT = (perSchoolPassingMath / per_school_counts) * 100
perSchoolPassingReadingPCT = (perSchoolPassingReading / per_school_counts) * 100
perSchoolPassingPCT = (perSchoolPassing / per_school_counts) * 100
perSchoolPassingPCT
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame(
    {
        "School Type": school_types,
        "Total Students": per_school_counts,
        "Total School Budget": perSchoolBudget,
        "Per Student Budget": perSchoolCapita,
        "Average Math Score": perSchoolMathScores,
        "Average Reading Score": perSchoolReadingScores,
        "% Passing Math": perSchoolPassingMathPCT,
        "% Passing Reading": perSchoolPassingReadingPCT,
        "% Overall Passing": perSchoolPassingPCT
    }


)

# Formatting
per_school_summary["Total Students"] = per_school_summary["Total Students"].map("{:,}".format)
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary["Average Math Score"] = per_school_summary["Average Math Score"].map("{:,.2f}".format)
per_school_summary["Average Reading Score"] = per_school_summary["Average Reading Score"].map("{:,.2f}".format)
per_school_summary["% Passing Math"] = per_school_summary["% Passing Math"].map("{:,.2f}%".format)
per_school_summary["% Passing Reading"] = per_school_summary["% Passing Reading"].map("{:,.2f}%".format)
per_school_summary["% Overall Passing"] = per_school_summary["% Overall Passing"].map("{:,.2f}%".format)

# Display the DataFrame
per_school_summary
```

## Scores by School Spending

```python
# display the per_school_summary dataframe
per_school_summary.head()
```

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary for later aggregations
school_spending_df = per_school_summary.copy()
school_spending_df
```

```python
# Use `pd.cut` on the per_school_capita Series from earlier to categorize per student spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(perSchoolCapita, bins=spending_bins, labels=labels)
school_spending_df
```

```python
# since mapping was used earlier, convert all of the following formatted columns back to floats
school_spending_df["Average Math Score"] = school_spending_df["Average Math Score"].astype(float)
school_spending_df["Average Reading Score"] = school_spending_df["Average Reading Score"].astype(float)
school_spending_df["% Passing Math"] = school_spending_df["% Passing Math"].str.replace("%","").astype(float)
school_spending_df["% Passing Reading"] = school_spending_df["% Passing Reading"].str.replace("%","").astype(float)
school_spending_df["% Overall Passing"] = school_spending_df["% Overall Passing"].str.replace("%","").astype(float)
school_spending_df
```

```python
# first, gorupby the spending ranges
#spendingMathDF = school_spending_df.groupby("Spending Ranges (Per Student)").mean()

# take the aggregate
#spendingGroupedDF = spendingMathDF.mean()
#spendingGroupedDF

# grab each series
spendingMathSeries = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spendingReadingSeries = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spendingPassMathSeries = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spendingPassReadingSeries = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
spendingPassingSeries = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summaryDF= pd.DataFrame(
    {
        "Average Math Score": spendingMathSeries,
        "Average Reading Score": spendingReadingSeries,
        "% Passing Math": spendingPassMathSeries,
        "% Passing Reading": spendingPassReadingSeries,
        "% Overall Passing": spendingPassingSeries
        
    }

)

spending_summaryDF["Average Math Score"] = spending_summaryDF["Average Math Score"].map("{:.2f}".format)
spending_summaryDF["Average Reading Score"] = spending_summaryDF["Average Reading Score"].map("{:.2f}".format)
spending_summaryDF["% Passing Math"] = spending_summaryDF["% Passing Math"].map("{:.2f}%".format)
spending_summaryDF["% Passing Reading"] = spending_summaryDF["% Passing Reading"].map("{:.2f}%".format)
spending_summaryDF["% Overall Passing"] = spending_summaryDF["% Overall Passing"].map("{:.2f}%".format)

# Display results
spending_summaryDF
```

## Scores by School Size

```python
# Create a copy of the school summary for later aggregations
schoolSpendingDF = school_spending_df.copy()
schoolSpendingDF
```

```python
# convert the total_students column
schoolSpendingDF["Total Students"] = schoolSpendingDF["Total Students"].str.replace(",","").astype(int)
schoolSpendingDF
```

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Use `pd.cut` on the per_school_counts Series from earlier to categorize school size based on the bins.
schoolSpendingDF["School Size"] = pd.cut(schoolSpendingDF["Total Students"], size_bins, labels=labels)
schoolSpendingDF
```

```python
# Calculate averages for the desired columns.
size_math_scores = schoolSpendingDF.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = schoolSpendingDF.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = schoolSpendingDF.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = schoolSpendingDF.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = schoolSpendingDF.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame(
    {
        "Average Math Score": size_math_scores,
        "Average Reading Score": size_reading_scores,
        "% Passing Math": size_passing_math,
        "% Passing Reading": size_passing_reading,
        "% Overall Passing": size_overall_passing
        

    }
    
)



# Display results
size_summary
```
